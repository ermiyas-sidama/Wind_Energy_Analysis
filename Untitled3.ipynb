```python
def tokenize_and_preprocess(reviews):
    stop_words = stopwords.words('english')
    stop_words.extend(['http','sxsw','hr'])
    patt = re.compile(r'\b(' + r'|'.join(stop_words) + r')\b\s+') 

    def preprocess_text(text):
        if pd.isna(text):  # Check for NaN values
            return ''  # Return empty string for missing values
        text = text.lower()  # Convert text to lowercase
        text = re.sub(r'[0-9]+', '', text)  # Remove numbers
        text = re.sub(patt, '', text)  # Remove stopwords
        return text
    
    preprocessed_text = reviews.apply(preprocess_text)  # Apply preprocessing function
    
    def remove_punct_and_stem(doc_tokenized):
        stemmer = nltk.WordNetLemmatizer()
        filtered_stemmed_tok = [stemmer.lemmatize(tok) for tok in doc_tokenized if tok.isalpha()]
        return " ".join(filtered_stemmed_tok)
        
    preprocessed = preprocessed_text.apply(word_tokenize).apply(remove_punct_and_stem)
        
    return preprocessed

df['tweet_text'] =tokenize_and_preprocess(df['tweet_text'])
```
